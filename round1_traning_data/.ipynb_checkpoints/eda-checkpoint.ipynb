```python
import json
train=[json.loads(k) for k in open("train.json","r",encoding="utf8").readlines()]
test=[json.loads(k) for k in open("testa.json","r",encoding="utf8").readlines()]
```

```python
import pandas as pd
train=pd.DataFrame(train)
test=pd.DataFrame(test)
```

```python
train.head()
```

```python
new_train=[]
sum=0
for unit in train.values:
    sum+=len(unit[2])
    assert len(unit[2])==len(unit[3])
    for event,summarization in zip(unit[2],unit[3]):
        assert event["id"] == summarization["id"]
        new_train.append([unit[0],unit[1],event["content"],summarization['event-summarization']])
    
        
        
```

```python
len(new_train),sum
```

```python
new_train=pd.DataFrame(new_train)
```
