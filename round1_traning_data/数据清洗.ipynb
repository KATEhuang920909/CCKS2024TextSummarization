```python
#
import json
train=[json.loads(k) for k in open("train.json","r",encoding="utf8").readlines()]
```

```python
content_summary=[]
for unit in train:
    for content,summary in zip(unit['events'],unit['summarizations']):
        assert content["id"]==summary["id"]
        if content["content"].strip()!="" and summary["event-summarization"].strip()!="":
            content_summary.append([content["content"],summary["event-summarization"]])
```

```python
len(content_summary)
```

```python
train[0]
```

```python
content_summary_5=[]
for unit in train:
    for content,summary in zip(unit['events'],unit['summarizations']):
        assert content["id"]==summary["id"]
        if content["content"].strip()!="" and summary["event-summarization"].strip()!="":
            content_final=""
            for span in content["spans"]:
                if span["end"]-span["start"]>=4:
                    content_final+=unit["doc"][span["start"]:span["end"]]
                else:
                    print(span,content["content"])
            if content_final.strip()!="":
                content_summary_5.append([content_final,summary["event-summarization"]])
            
#             
```

```python
len(content_summary_5),len(content_summary)
```

```python
import pandas as pd
content_summary=pd.DataFrame(content_summary)
content_summary_5=pd.DataFrame(content_summary_5)
```

```python
content_summary = content_summary.drop_duplicates()
content_summary_5 = content_summary_5.drop_duplicates()
```

```python
content_summary=content_summary.values.tolist()
content_summary_5=content_summary_5.values.tolist()
```

```python
content_summary=[{"content":k[0],"summarization":k[1]} for k in content_summary]
content_summary_5=[{"content":k[0],"summarization":k[1]} for k in content_summary_5]
from random import shuffle
shuffle(content_summary),shuffle(content_summary_5)
train=content_summary[:int(0.8*len(content_summary))]
test=content_summary[int(0.8*len(content_summary)):]

train_5=content_summary_5[:int(0.8*len(content_summary_5))]
test_5=content_summary_5[int(0.8*len(content_summary_5)):]
```

```python
len(train),len(test),len(train_5),len(test_5)
```

```python
# 小于50的用原文
import os
if not os.path.exists("data_clean/"):
    os.mkdir("data_clean/")
with open('data_clean/train.txt', 'w',encoding="utf8") as file:
    for unit in train:
        json_str = json.dumps(unit,ensure_ascii=False)
        file.write(json_str+"\n")
with open('data_clean/test.txt', 'w',encoding="utf8") as file:
    for unit in test:
        json_str = json.dumps(unit,ensure_ascii=False)
        file.write(json_str+"\n")
with open('data_clean/train_5.txt', 'w',encoding="utf8") as file:
    for unit in train_5:
        json_str = json.dumps(unit,ensure_ascii=False)
        file.write(json_str+"\n")
with open('data_clean/test_5.txt', 'w',encoding="utf8") as file:
    for unit in test_5:
        json_str = json.dumps(unit,ensure_ascii=False)
        file.write(json_str+"\n")
```
